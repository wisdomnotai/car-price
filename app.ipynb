## Data Cleaning and preparation for modelling





```python
#importing necessary libaries

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#loading the dataset

car_price = pd.read_csv("dataset/data.csv")

#displaying top 5 and bottom 5 rows

car_price.head()
car_price.tail()
```

```python
#checking the datatypes in the dataset

car_price.dtypes
```

```python
#checking for irrelevant columns
car_price.info()
```

```python
#dropping irrelevant columns 

car_price = car_price.drop(["Engine Fuel Type", "Market Category", "Vehicle Style", "Popularity", "Number of Doors", "Vehicle Size"], axis = 1)
car_price.info()
```

```python
#Renaming Columns

car_price = car_price.rename(columns = {"Engine HP":"HP", "Engine Cylinders":"Cylinders", "Transmission Type": "Transmission",
                             "Driven Wheels":"Drive Mode", "Highway MPG":"MPG-H", "City mpg":"MPG-C", "MSRP":"Price"})
car_price.info()
```

```python
#Droppping Dupplicate rows

car_price.shape
car_price.count()
```

```python
#Finding the number of duplicated data
car_price.duplicated().sum()
```

```python
#Dropping duplicated data
car_price = car_price.drop_duplicates()
car_price.count()
```

```python
#checking for missing values

car_price.isnull().sum()
```

```python
#only a few null values so they will be dropped 

car_price = car_price.dropna()
car_price.isnull().sum()
```
